```python
## Now let's try our projection tricks. This was an earlier attempt where I projected
## semi-axes and rotated them etc

def projection(index1, index2,P): #index1 and index2 are the indices to pick from the D matrix
    # axis1 and axis2 are the pair of axes to project on e.g. (0,1) : (x,y) plane
    sigma1 = np.sqrt(1/D[index1,index1])
    sigma2 = np.sqrt(1/D[index2,index2])
    
    ubasis_1 = np.linspace(-sigma1,sigma1,100)
    ubasis_2 = sigma2*np.sqrt(1 - ubasis_1**2/sigma1**2)
    
    #plt.scatter(ubasis_1,ubasis_2)
    #plt.scatter(ubasis_1,-ubasis_2)
    #plt.show()
    
    #ubasis_3 = np.zeros(len(ubasis_1))
    
    u_mat = np.zeros((3,len(ubasis_1)))
    u_mat[index1,:] = ubasis_1
    u_mat[index2,:] = ubasis_2
    #the rest of the u_matrix is just zero.
    #u_mat[2,:] = ubasis_3
    
    #branch2
    u_mat2 = np.zeros((3,len(ubasis_1)))
    u_mat2[index1,:] = ubasis_1
    u_mat2[index2,:] = -ubasis_2
    #u_mat2[2,:] = ubasis_3
    
    ##rotation
    x_mat = np.matmul(P, u_mat)
    x_mat2 = np.matmul(P, u_mat2)
    
    return x_mat, x_mat2
```

```python
import sympy
def rref(matrix):
  """
  Computes the reduced row echelon form (RREF) of a matrix.

  Args:
    matrix: A NumPy array representing the matrix.

  Returns:
    A tuple containing:
      - The reduced row echelon form of the matrix (as a sympy Matrix object).
      - A tuple of pivot column indices.
  """
  matrix = sympy.Matrix(matrix)
  return matrix.rref()

```

```python
def cyclically_permute_columns(arr, shift):
    """
    Cyclically permutes the columns of a NumPy array.

    Args:
        arr (np.ndarray): The input NumPy array.
        shift (int): The number of positions to shift the columns. 
                     Positive shift values move columns to the right, 
                     negative values move them to the left.

    Returns:
        np.ndarray: A new array with cyclically permuted columns.
    """
    num_cols = arr.shape[1]
    shift = shift % num_cols  # Handle shifts larger than the number of columns
    
    if shift == 0:
      return arr.copy()
    
    return np.concatenate((arr[:, -shift:], arr[:, :-shift]), axis=1)

# Example usage

shift_amount = 2
permuted_arr = cyclically_permute_columns(B, shift_amount)
print(permuted_arr)
```

```python
## Ok I tried a lot of things. Now let's write a proper function that can do the projection on x1,x2 plane to start with.
def project_xy(A, axis1,axis2,numpoints):
    
    dimension = np.shape(A)[0] # this is n
    
    B = np.delete((A + A.T),[axis1,axis2],axis=0)#(A + A.T)[2:,:] # selects only rows >= 3 onwards, because we want to project onto x1,x2 plane. B has shape n-2 , n
    # where n = dim(d) is the dimension of the model space
    # equation we want to solve is : B.[x1 x2 x3 x4 ... xn] = 0 
    # I want x3(x1,x2) , x4(x1,x2) , ..., xn(x1,x2). Due to the way reduced row echelon works, I need to cyclically permute the x1,x2 columns
    # of B to the end such that B_shifted.[x3 x4 ... xn x1 x2] = 0. Then row-reducing B_shifted will give what I want.
    
    
    B_shifted = permuted_mat(B, axis1,axis2)

    print("B is",B)
    print("B_shifted is", B_shifted)
    
    #create sympy vector of symbols of dimension n = dim(d).
    symbol_list = sympy.symbols(f'x1:{dimension+1}')
    vec = sympy.Matrix(symbol_list)
    print(vec)

    axis1_symbol = vec[axis1]
    axis2_symbol = vec[axis2]

    reduced_B_shifted, pivot_cols = rref(B_shifted)

    reduced_B_shifted_numpy = np.array(reduced_B_shifted).astype(np.float64)

    print(reduced_B_shifted_numpy)
    print("notshifted", rref(B))

    #so that I don't get an error when going through the for loop. I need to artificially augment reduced_B_shifted
    reduced_B_shifted_numpy = np.insert(reduced_B_shifted_numpy,[axis1,axis2-1],np.zeros((2,dimension)),axis=0)

    for i in range(dimension):
        if i == axis1 or i == axis2:
            pass
        else:
            vec[i] = (-1)*reduced_B_shifted_numpy[i,-2]*axis1_symbol + (-1)*reduced_B_shifted_numpy[i,-1]*axis2_symbol 
    print(vec)

    expr = sympy.simplify((vec.T*A*vec)[0])

    print(expr)

    # Now it is just a question of extracting coefficients to get the smaller 2by2 matrix for ellipse in x1,x2 plane.


    A_selected = np.zeros((2,2)) # project on 2d plane

    A_selected[0,0] = np.float64(expr.coeff(axis1_symbol,2))
    A_selected[0,1] = np.float64(expr.coeff(axis1_symbol*axis2_symbol,1))/2
    A_selected[1,0] = np.float64(expr.coeff(axis1_symbol*axis2_symbol,1))/2
    A_selected[1,1] = np.float64(expr.coeff(axis2_symbol,2))

    print(A_selected)

    #diagonalize A_selected to find A_selected^(1/2)
    eigval, eigvec = np.linalg.eig(A_selected)
    print(eigval)
    diag_mat = np.diag(eigval)
    A_selected_half = np.matmul(eigvec,np.matmul(np.sqrt(diag_mat), eigvec.T))
    

    dim_d = 2
    coordinates_d = np.random.uniform(-1,1,(numpoints,dim_d)) #needs to be changed to -1 in the actual code too.
    norms = np.diag(1/np.sqrt(np.sum(coordinates_d**2,axis=1)))
    coordinates_d = np.dot(norms,coordinates_d)
    

    val = np.einsum('ij,kj-> ik',np.linalg.inv(A_selected_half), coordinates_d).T  # this gives ellipsoid

    return val

    
        
    

    
```
